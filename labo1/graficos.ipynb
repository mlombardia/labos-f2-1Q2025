```python
import pandas as pd
import matplotlib.pyplot as plt

archivo_csv = 'datos.csv'  
df = pd.read_csv(archivo_csv)

columna_x = 'Voltaje'
columna_y = 'Corriente'

valores_voltaje = df['voltage']
corriente_teorica = valores_voltaje / 10
corriente_experimental = df['current']

resol_voltaje=0.01
percentage_voltaje=0.01
digits_voltaje=5

resol_corriente=0.01
percentage_corriente=0.02
digits_corriente=10

error_voltaje = percentage_voltaje * valores_voltaje + digits_voltaje * resol_voltaje
error_corriente = percentage_corriente * corriente_experimental + digits_corriente * resol_corriente
print(error_voltaje)
print(error_corriente)

plt.figure(figsize=(8, 5))
plt.errorbar(valores_voltaje, corriente_experimental, xerr=error_voltaje, yerr=error_corriente, fmt='o', capsize=5)
plt.plot(valores_voltaje, corriente_teorica, 'r-', label='Voltaje Teórico')
plt.scatter(valores_voltaje, corriente_experimental, marker='o', label='Datos Medidos')
plt.xlabel(columna_x + '(V)')
plt.ylabel(columna_y + '(mA)')
plt.title(f'{columna_x}(V) vs {columna_y}(mA) para resistencia de 10kΩ')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
