```python
import pandas as pd
import matplotlib.pyplot as plt

archivo_csv = 'datos.csv'  
df = pd.read_csv(archivo_csv)

columna_x = 'Voltaje'
columna_y = 'Corriente'

corriente_teorica = df['voltage'] / 10

plt.figure(figsize=(8, 5))
plt.errorbar(df['voltage'], df['current'], xerr=0.0225, yerr=0.1, fmt='o', capsize=5)
plt.plot(df['voltage'], corriente_teorica, 'r-', label='Voltaje Te√≥rico')
plt.scatter(df['voltage'], df['current'], marker='o', label='Datos Medidos')
plt.xlabel(columna_x + '(V)')
plt.ylabel(columna_y + '(mA)')
plt.title(f'{columna_x}(V) vs {columna_y}(mA) para resistencia de 10kŒ©')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
archivo_csv = 'datos2.csv'  
df = pd.read_csv(archivo_csv)

columna_x = 'Voltaje'
columna_y = 'Corriente'

plt.figure(figsize=(8, 5))
plt.errorbar(df['voltage'], df['current'], xerr=0.0121, yerr=1, fmt='o', capsize=5)
plt.plot(df['voltage'], df['voltage'], 'r-', label='Voltaje Te√≥rico')
plt.scatter(df['voltage'], df['current'], marker='o', label='Datos Medidos')
plt.xlabel(columna_x + '(V)')
plt.ylabel(columna_y + '(mA)')
plt.title(f'{columna_x}(V) vs {columna_y}(mA) para resistencia de 1kŒ©')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
