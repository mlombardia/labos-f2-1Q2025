```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV
archivo_csv = 'datos.csv'  # Reemplazá con la ruta real de tu archivo
df = pd.read_csv(archivo_csv)

# Seleccionar las columnas que querés graficar
columna_x = 'voltage'  # Reemplazá con el nombre de la columna para el eje X
columna_y = 'current'  # Reemplazá con el nombre de la columna para el eje Y

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(df[columna_x], df[columna_y], marker='o')
plt.xlabel(columna_x)
plt.ylabel(columna_y)
plt.title(f'{columna_x} vs {columna_y}')
plt.grid(True)
plt.tight_layout()
plt.show()
```
